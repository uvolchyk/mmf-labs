```python
def critical(matrix):
    i=j=0
    time = matrix[0][0]
    
    while i < len(matrix) - 1:
        while j < len(matrix) - 1:
            if matrix[i][j+1] > matrix[i+1][j]:
                j+=1
                time += matrix[i][j]
                continue
            else:
                time += matrix[i+1][j]
                break
        i+=1
    return time
```
